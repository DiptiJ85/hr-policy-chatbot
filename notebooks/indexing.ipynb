```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="../env/.env")
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import os
import shutil
```

```python
#Clear ChromaDB before begining
if os.path.exists("../chroma_db"):
    shutil.rmtree("../chroma_db")
    print("Cleared existing ChromaDB")
else:
    print("Nothing to clear")

# Load PDF document
loader = PyPDFLoader(file_path="../Dataset/the_nestle_hr_policy_pdf_2012.pdf")
documents = loader.load()
print(f"Loaded {len(documents)} pages")

# Split PFD document into Chunks of 500 size
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
chunks = splitter.split_documents(documents)
print(f"Split into {len(chunks)} chunks")

# Embedding Model
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")

# Create vectors and store into ChromaDB
vector_store = Chroma.from_documents(documents=chunks, embedding=embeddings, persist_directory="../chroma_db")
print(f"Stored {len(chunks)} chunks in ChromaDB")

print("Indexing complete !")
```
